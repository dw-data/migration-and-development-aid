```python
#sort the data first by recipient and then by donor and year
ODAfromall=ODAfromall.sort_values(by=["Recipient","Donor","Year"])
testdata=ODAfromall
#testdata=testdata.head()
testdata
```

```python
#create an empty dataframe for storing the results
emptyframe = pd.DataFrame(columns=["Recipient","Year","Money spent per capita"])

#iterate through the dataframe row by row
for index, row in testdata.iterrows():
    #if the donor is "all donors" and if the aid type is the oda total, gross
    if row["Donor"]=="All Donors, Total" and row["Aid type"]=="Memo: ODA Total, Gross disbursements" and row["Amount type"]=="Constant Prices":
        #then set the recipient variable to recipient of the row
        recipient=row["Recipient"]
        annus=str(row["Year"])
        try:
            #we divide the foreign aid transferred to this country by the population of this country
            cashpercap=row["Value"]*1000000/(int(population.loc[recipient][annus].replace(" ", ""))*1000)
            cash=row["Value"]*1000000
            #and append the name of the recipient, the year and the sum spent per capita
            df2 = pd.DataFrame([[recipient,annus,cash,cashpercap]], columns=["Recipient","Year","Overall Sum","Money spent per capita"])
            emptyframe=emptyframe.append(df2,ignore_index=True)
        except KeyError:
            continue
        
```

```python
#sanity checks
emptyframe=emptyframe[["Recipient","Year","Overall Sum","Money spent per capita"]]
print(emptyframe.head())
```

```python
#dropping all NaNs
print(len(emptyframe))
emptyframe=emptyframe.dropna(axis=1, how='any')
print(len(emptyframe))
```

```python
#now we run this script to calculate the average over three years as described in the paper
#first creating an emptydataframe to append the results of our calculations to
emptyframe3 = pd.DataFrame(columns=['Recipient','Report Year',"3yrAverage Money"])

#now we define for each column two variables that allow us to track
#the recipient, the year and the money from the previous year
#and the recipient, the year and the money from the year before 
previousrecipient="0"
recipient2="0"

previousyear=0
year2=0

previousmoney=0
money2=0

previousmoneypercap=0
moneypercap2=0

#now we iterate through the dataframe
for index, row in testdata.iterrows():
    #and pick up the recipient, the year and the money received for the current row
    recipient=row[0]
    year=int(row[1])
    money=row[2]
    moneypercap=row[3]
    
    #if the number of current year is bigger than from the previous year
    #and the previous year is one year bigger than the year before
    if (year-1==previousyear)and(previousyear-1==year2):
        #and the recipient has been the same through all of these years
        if (recipient==previousrecipient)and(previousrecipient==recipient2):
            
            #we calculate the average sum of devaid built from the current year, the previous year and the year before
            average=(money+previousmoney+money2)/3
            averagepercap=(moneypercap+previousmoneypercap+moneypercap2)/3
            #and define the reportyear as the year after
            reportyear=int(year+1)
            
            #and then we add the recipient, the reportyear and the average of these three years to our dataframe
            df2 = pd.DataFrame([[recipient,reportyear,average, averagepercap]], columns=['Recipient','Report Year',"3yrAverage Money","3yrAverage Money per cap"])
            emptyframe3=emptyframe3.append(df2,ignore_index=True)
            
            #and then we redefine for these variable by shifting them
            #the recipient from two years ago becomes the one from the previous year
            #the previous recipient becomes the one from the current row
            #and so on
            recipient2=previousrecipient
            previousrecipient=recipient

            year2=previousyear
            previousyear=year

            money2=previousmoney
            previousmoney=money
            
            moneypercap2=previousmoneypercap
            previousmoneypercap=moneypercap
        else:
            pass
    
    else:
        pass
        #and we do this shift/exchange also when the if statement was not true
        recipient2=previousrecipient
        previousrecipient=recipient

        year2=previousyear
        previousyear=year

        moneypercap2=previousmoneypercap
        previousmoneypercap=moneypercap
        
        
```

```python
#again we define an empty dataframe
emptyframe2 = pd.DataFrame(columns=['Country of Birth',"Destination","Emigration sum",'Emigration per cap'])
            
counter=0
#and now we iterate through the dataframe
for index, row in testdata.iterrows():
    counter+=1
    #this is just a simple progress counter
    if (counter%2000==0):
        print(counter)
    #only when the variable is about the inflow we process the data
    if row[3]=="Inflows of foreign population by nationality":
        #we set the birthcountry variable on the "Country of birth/nationality" value of the row and annus to the row's year
        birthcountry=row[1]
        annus=str(row["Year"])
        #we set the migdestination variable on the "Country" value of the row, so the country that has taken up the migrants
        migdestination=row["Country"]
        try:
            #now we dividee the migration number by the population
            mig=row["Value"]
            migpercap=row["Value"]/(int(population.loc[birthcountry][annus].replace(" ", ""))*1000)
            #and append all three variables to the dataframe
            df2 = pd.DataFrame([[birthcountry,annus, migdestination,mig,migpercap]], columns=['Country of Birth',"Year","Destination","Emigration sum",'Emigration per cap'])
            emptyframe2=emptyframe2.append(df2,ignore_index=True)
       
        except KeyError:
                pass
  
```

```python
#now we use this newly created dataframe to calculate the overall outflow of migrants from any country into OECD countries
testdata=emptyframe2
testdata=testdata.sort_values(by=["Country of Birth","Year"])
testdata.head()
#testdata=testdata.head(1000)
```

```python
#again we create an emptydataframe
emptyframe3 = pd.DataFrame(columns=['Country of Birth','Emigration per cap over all OECD','Emigration sum over all OECD countries',"Year"])
#we set the variables to default values
previousbirth="0"
previousannus="0"
sumOECD=0
sumOECDpercap=0
counter=0
#we iterate through the dataframe
for index, row in testdata.iterrows():
    counter+=1
    if (counter%2000==0):
        print(counter)
    #in each row the country of birth variable is taken from the rows 0th column
    birthcountry=row[0]
    #and the row's year
    annus=row[4]
    #if the previous birth country was not the default value and is not the current birth country
    #or the year is not the year of this row is different from the row before
    if (previousbirth != "0" and previousbirth!=birthcountry)or(previousannus != "0" and previousannus!=annus):
        #we then append the country of birth, the year and the sum of emigration to the dataframe
        df2 = pd.DataFrame([[previousbirth,sumOECDpercap,sumOECD,previousannus]], columns=['Country of Birth','Emigration per cap over all OECD','Emigration sum over all OECD countries',"Year"])
        emptyframe3=emptyframe3.append(df2,ignore_index=True)
        #we reset the number of migrants to zero and then add the value of the current row
        sumOECD=0
        sumOECD+=row["Emigration sum"]
        sumOECDpercap=0
        sumOECDpercap+=row["Emigration per cap"]
        #and we change the previousbirth variable to the one of the current country
        #and the previous year to the one of the current yrow
        previousbirth=birthcountry
        previousannus=annus
    #if the previous row and the current row have the same birthcountry then
    else:
        #we keep adding the value of emigration to the sum
        sumOECD+=row["Emigration sum"]
        sumOECDpercap+=row["Emigration per cap"]
        #and we also change the previousbirth variable to the one of the current country
        previousbirth=birthcountry
        previousannus=annus
        
        
        
```

```python
#and now we also calculate three year averages
emptyframe3 = pd.DataFrame(columns=['Recipient','Report Year',"3yrAverage Migration per Cap","3yrAverage Migration"])
#We set up some variables that should store what was in the row before and the row even before that
previousrecipient="0"
recipient2="0"

previousyear=0
year2=0

previousmig=0
mig2=0

previousmigcap=0
migcap2=0

#we iteratre through the dataframe
for index, row in testdata.iterrows():
    #and get the recipient, the year and the number of migrants per capita
    recipient=row[0]
    year=int(row[3])
    migcap=row[1]
    mig=row[2]
    
    #we check whether we have three continous years
    #and the same country
    if (year-1==previousyear)and(previousyear-1==year2):
        if (recipient==previousrecipient)and(previousrecipient==recipient2):
            #then we calculate the 3year-average
            averagecap=(migcap+previousmigcap+migcap2)/3
            average=(mig+previousmig+mig2)/3
            #and add for the report year 1 year
            reportyear=int(year+1)
            #then we append the variables to the dataframe
            df2 = pd.DataFrame([[recipient,reportyear,averagecap,average]], columns=['Recipient','Report Year',"3yrAverage Migration per Cap","3yrAverage Migration"])
            emptyframe3=emptyframe3.append(df2,ignore_index=True)
            
            #we shift all variables by one row/year
            recipient2=previousrecipient
            previousrecipient=recipient

            year2=previousyear
            previousyear=year
            
            migcap2=previousmigcap
            previousmigcap=migcap
            
            mig2=previousmig
            previousmig=mig
        else:
            pass
    
    else:
        pass
        #we shift all variables by one row/year

        recipient2=previousrecipient
        previousrecipient=recipient

        year2=previousyear
        previousyear=year
        
        migcap2=previousmigcap
        previousmigcap=migcap

        mig2=previousmig
        previousmig=mig
        
        
```
